# GasTurbine CO and NOx Emission DataSet
in this project we want to predict turbine energy yield (TEY) using ambient variables as features. we have to find a good model for predicting the TEY.we will use some algorithms to find best model: LinearRegression and Lasso and Ridge.

# importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from pandas.plotting import scatter_matrix
from sklearn.metrics import r2_score
```

# reading dataset and obtaining some information about it

```python
files=["gt_2011.csv","gt_2012.csv","gt_2013.csv","gt_2014.csv","gt_2015.csv"]
df=pd.DataFrame()
for file in files:
    data=pd.read_csv(file)
    df=pd.concat([df,data],axis=0)

df.to_csv("merged_files.csv",index=False)
```

```python
df.info()
```

```python
df.describe()
```

# finding correlation between target(TEY) and other features

```python
corr=df.corr()
sb.heatmap(corr,xticklabels=corr.columns,yticklabels=corr.columns,annot=True)
```

```python
mycopy=df.copy()
```

# plotting

```python
features=["AFDP","GTEP","TIT","CDP","TEY"]
scatter_matrix(mycopy[features],figsize=(10,15))
plt.show()
```

```python
X=mycopy[["AFDP","GTEP","TIT","CDP"]].values
Y=mycopy["TEY"].values
```

# train_test_split

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=5)
```

# LinearRegression

```python
model1=LinearRegression()
myfit1=model1.fit(x_train,y_train)
```

```python
pred1=model1.predict(x_test)
```

```python
score=r2_score(y_test,pred1)
mean=np.sum((pred1-y_test)**2)/36733
print(f"R2-score : {score}")
print(f"error : {mean}")
```

as you see linearRegression had awesome prediction but it is possible that maybe the model has overfitting so for exploring this subject we can use regularization methods like Lasso and Ridge.

# Lasso

```python
model2=Lasso(alpha=0.01)
model2.fit(x_train,y_train)
```

```python
pred2=model2.predict(x_test)
```

```python
score=r2_score(y_test,pred2)
mean=np.sum((pred2-y_test)**2)/36733
print(f"R2-score : {score}")
print(f"error : {mean}")
```

# Ridge

```python
model3=Ridge(alpha=0.01)
model3.fit(x_train,y_train)
```

```python
pred3=model3.predict(x_test)
```

```python
score=r2_score(y_test,pred3)
mean=np.sum((pred3-y_test)**2)/36733
print(f"R2-score : {score}")
print(f"error : {mean}")
```

as you see these three models don't differ much. their prediction accuracy is close to each other with slight difference. you can see that Lasso has relatively less error than others.
